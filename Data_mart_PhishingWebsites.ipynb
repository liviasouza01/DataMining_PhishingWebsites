```python
import pandas as pd

url = 'https://raw.githubusercontent.com/liviasouza01/DataMining_PhishingWebsites/main/url_data.csv'
df = pd.read_csv(url)

df.head()
```

```python
print(df.isnull().sum())
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
import sqlite3

conn = sqlite3.connect('phishing_websites_datamart.db') #conectar o banco de dados

df.to_sql('phishing_websites', conn, if_exists='replace', index=False)

conn.close() # Fechar a conexão


print("Dados armazenados no banco de dados SQLite com sucesso!")
```

###Consulta e análise

```python
conn = sqlite3.connect('phishing_websites_datamart.db')

# Fazer uma consulta SQL
query = "SELECT * FROM phishing_websites WHERE Label = 'phishing'"
phishing_sites = pd.read_sql_query(query, conn)

# Exibir os resultados da consulta
phishing_sites.head()
```
